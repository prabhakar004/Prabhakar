```

from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("DataFrame").getOrCreate()

df = spark.read.option("header","true").csv("list_of_cities_and_towns_in_india-834j.csv")

df.show(4,False)
```

```
dataCollect = df.collect()
```

```
print(dataCollect)
```

```
dfcollect2= df.select("Name of City").collect()
print(dfcollect2)
```

```
for row in dataCollect:
  if row['Name of City'] !=None or row["State"] !=None:
    print (row['Name of City']+","+str(row["State"]))
```
